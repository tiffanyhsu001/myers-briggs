### Neural Network Prediction Model
This notebook breaks down the myers briggs personality types into its 4 subcategories. Each of the subcategories is broken down into its own NN predictive model.

https://towardsdatascience.com/optimizing-neural-networks-where-to-start-5a2ed38c8345

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import tensorflow as tf
from nn import neural_net

import warnings
warnings.filterwarnings('ignore')

import os
import sys
dir = os.getcwd()
data_path = os.path.sep.join(dir.split(os.path.sep)[:-1])
```

```python
mbti = pd.read_csv(data_path + '/mbti_updated.csv', index_col=0)

x = mbti[['words_per_comment', 'var_word_cnts', 'question_marks', 'exclam_marks', 'ellipses_cnt',
       'links_cnt', 'afinn_avg', 'noun_avg', 'verb_avg',
       'polarity_avg', 'subj_avg']]
```

### Introvert vs Extrovert
Where 1 = introvert, 0 = extrovert

```python
I_E = mbti['I_E']
```

```python
nn = neural_net(x, I_E)
```

```python
nn.train_test_split()
```

```python
nn.oversample('I','E')
```

```python
learning_rate = 0.1
epochs = 1000
decay_rate = learning_rate / epochs
momentum = 0.8

nn.sgd(learning_rate, decay_rate, momentum)
```

```python
nn.build_model()
nn.compile()

batch_size= 256
nn.fit(batch_size, epochs, 'I') # I = 1, E = 0
```

```python
nn.plot_loss()
```

```python
nn.plot_acc()
```

```python
nn.plot_f1()
```

```python
nn.predict()
```
