### Myers Briggs Personality Classification - Exploratory Analysis & Feature Engineering
The objective of this project is to predict/classify personality types based on text forum posts. Each personality type consits of FOUR types:
    1. I/E: introverts vs. extroverts
    2. N/S: intuitives vs. sensers
    3. T/F: thinkers vs. feelers
    4. J/P: judgers vs. perceivers
Each subcategory will be classified individually, then combined together for an ultimate overall score.


This notebook consists of initial exploratory analysis: checking for personality type imbalances prior to classification. This notebook also includes feature engineering to break down text posts.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk import word_tokenize, pos_tag
from afinn import Afinn
from textblob import TextBlob
from scipy.stats import chi2_contingency
from scipy.stats import chi2
from itertools import combinations


import warnings
warnings.filterwarnings('ignore')
```

```python
mbti = pd.read_csv('mbti_1.csv')
```

```python
mbti['type'].value_counts().plot(kind='bar', figsize=(12,8),color='lightseagreen')
```

Clearly, there are some class inbalances here, let's look closer into each subcategory.

```python
# Finding the number of each personality type, broken down by category
mbti['I_E'] = 0
mbti['N_S'] = 0
mbti['F_T'] = 0
mbti['J_P'] = 0

for i in range(mbti['type'].shape[0]):
    if "I" in mbti['type'][i][0]:
         mbti['I_E'][i] = 'I'
    elif 'E' in mbti['type'][i][0]:
        mbti['I_E'][i] = 'E'
    if "N" in mbti['type'][i][1]:
         mbti['N_S'][i] = 'N'
    elif 'S' in mbti['type'][i][1]:
        mbti['N_S'][i] = 'S'
    if "F" in mbti['type'][i][2]:
         mbti['F_T'][i] = 'F'
    elif 'T' in mbti['type'][i][2]:
        mbti['F_T'][i] = 'T'
    if "J" in mbti['type'][i][3]:
         mbti['J_P'][i] = 'J'
    elif 'P' in mbti['type'][i][3]:
        mbti['J_P'][i] = 'P'
```

### Internal Analysis - Testing for Independence Amongst The 4 Subcategories
To get a better understanding of the four subcategories, we will perform Chi Sq Test for Independence to see if certain subcategory types are more likely to be one or the other based off another subcategory.

Example: 
Are Introverts more likely to be Intuitive or Sensing? 
Are Extroverts more likely to be Intuitive or Sensing? 

```python
def chi_sq(cat1,cat2):
    table = pd.crosstab(mbti[cat1], mbti[cat2],  margins = False) 
    stat, p, dof, expected = chi2_contingency(table)
    
    critical = chi2.ppf(0.95, dof)
    if abs(stat) >= critical:
        print(cat1, 'and', cat2,'are dependent (reject H0)')
    else:
        print(cat1, 'and', cat2,'are independent (fail to reject H0)')
```

```python
for combo in combinations(['I_E', 'N_S', 'F_T', 'J_P'], 2):  # checking independence amongst all combos of pairs
    chi_sq(combo[0], combo[1])
```

#### Chi Sq Results
Introverted/Extroverted traits are dependent on Intuitive/Sensing traits, meaning perhaps an introverted person is more likely to be intuitive or vice versa. Similarly, Introverted/Extroverted traits are also dependent on Feeling/Thinking and Judging/Perceiving. Intuitive/Sensing traits are also dependent on Judging/Perceiving traits.

Intuition/Sensing and Judging/Perceiving are independent; Feeling/Thinking and Judging/perceiving are also independent.

## Feature Engineering
1. number of exclamation marks, question marks, ellipses, words per comment, links per comment
2. number of nouns, verbs
3. avg afinn sentiment analysis score per comment
4. ratio of bing positive to neg analysis score

```python
mbti['question_marks'] = mbti['posts'].apply(lambda x: x.count('?')/50)
mbti['exclam_marks'] = mbti['posts'].apply(lambda x: x.count('!')/50)
mbti['ellipses_cnt'] = mbti['posts'].apply(lambda x: x.count('...')/50)
mbti['links_cnt'] = mbti['posts'].apply(lambda x: x.count('http://')/50)
```

```python
def var_row(row):
    l = []
    for i in row.split('|||'):
        l.append(len(i.split()))
    return np.var(l)
```

```python
mbti['words_per_comment'] = mbti['posts'].apply(lambda x: len(x.split())/50)
mbti['var_word_cnts'] = mbti['posts'].apply(lambda x: var_row(x))
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot("type", "words_per_comment", data=mbti)
plt.title('Type vs Avg Words Per Comment')
```

#### Counting Parts of Speech

```python
def noun_avg(x):
    x.replace('|||', ' ')
    noun_avg = sum(1 for word, pos in pos_tag(word_tokenize(x)) if pos.startswith('NN'))/50
    return noun_avg

def verb_avg(x):
    x.replace('|||', ' ')
    verb_avg = sum(1 for word, pos in pos_tag(word_tokenize(x)) if pos.startswith('VB'))/50
    return verb_avg
```

```python
mbti['noun_avg'] = mbti['posts'].apply(lambda x: noun_avg(x))
```

```python
mbti['verb_avg'] = mbti['posts'].apply(lambda x: verb_avg(x))
```

#### AFINN Sentiment Analysis
AFINN looks up words from a lexicon with preset values from -5 to 5. Words like happy will have a value of 5, while death will have a value of -5. AFINN sentiment analysis does not account for negation terms (i.e 'I am NOT happy'), it only looks at individual words.

```python
af = Afinn()
```

```python
mbti['posts_combined'] = mbti['posts'].apply(lambda x: x.replace('|||', ' '))
mbti['afinn_avg'] = mbti['posts_combined'].apply(lambda x: af.score(x)/50)
```

#### TextBlob Polarity & Subjectivity

```python
def polarity(x):
    blob = TextBlob(x)
    polarity = blob.sentiment[0]
    return polarity

def sub(x):
    blob = TextBlob(x)
    sub = blob.sentiment[1]
    return sub
```

```python
mbti['polarity_avg'] = mbti['posts_combined'].apply(lambda x: polarity(x))
mbti['subj_avg'] = mbti['posts_combined'].apply(lambda x: sub(x))
```
